```python
DATA_DIR= '/content/'
data_dir= pathlib.Path(DATA_DIR)
print(data_dir)
```

```python
train_ds, valid_ds= tf.keras.utils.audio_dataset_from_directory(
    directory= data_dir,
    batch_size=64,
    validation_split= 0.15,
    seed=0,
    output_sequence_length=16000,
    subset= 'both'
)
```

```python
label_names= np.array(train_ds.class_names)
print(label_names)
```

```python
def squeeze(audio, labels):
  audio= tf.squeeze(audio, axis=-1)
  return audio, labels
```

```python
train_ds= train_ds.map(squeeze, tf.data.AUTOTUNE)
valid_ds= valid_ds.map(squeeze, tf.data.AUTOTUNE)
```

```python
test_ds= valid_ds.shard(num_shards=2, index=0)
valid_ds= valid_ds.shard(num_shards=2, index=1)
```

```python
def get_spectogram(waveform):
  spectogram= tf.signal.stft(waveform, frame_length=255, frame_step=128)
  spectogram= tf.abs(spectogram)

  spectogram= spectogram[..., tf.newaxis]
  return spectogram
```

```python
def make_spec_ds(ds):
  return ds.map(
      map_func= lambda audio, label:(get_spectogram(audio), label),
      num_parallel_calls= tf.data.AUTOTUNE
  )
```

```python
train_spec_ds= make_spec_ds(train_ds)
test_spec_ds= make_spec_ds(test_ds)
valid_spec_ds= make_spec_ds(valid_ds)
```

```python
train_spec_ds= train_spec_ds.cache().shuffle(10000).prefetch(tf.data.AUTOTUNE)
test_spec_ds= test_spec_ds.cache().prefetch(tf.data.AUTOTUNE)
valid_spec_ds= valid_spec_ds.cache().prefetch(tf.data.AUTOTUNE)
```

```python
for example_spectograms, example_labels in train_spec_ds.take(1):
  break;
```

```python
shape_input= example_spectograms.shape[1:]
num_labels= len(label_names)
norm_layer= layers.Normalization()
norm_layer.adapt(train_spec_ds.map(map_func= lambda spec, label:spec))

model= models.Sequential([
    layers.Input(shape= shape_input),
    layers.Resizing(32,32),
    norm_layer,
    layers.Conv2D(32,3, activation= 'relu', padding= 'same'),
    layers.MaxPool2D(),
    layers.Conv2D(64,3, activation= 'relu', padding= 'same'),
    layers.MaxPool2D(),
    layers.Conv2D(128,3, activation= 'relu', padding= 'same'),
    layers.MaxPool2D(),
    layers.Dropout(0.5),
    layers.Flatten(),
    layers.Dense(64),
    layers.Dropout(0.25),
    layers.Dense(num_labels)


])

model.summary()
```

```python
model.compile(
    optimizer= tf.optimizers.Adam(),
    loss= tf.losses.SparseCategoricalCrossentropy(from_logits= True),
    metrics= ['accuracy']
)
```

```python
EPOCHS= 3

history= model.fit(
    train_spec_ds,
    validation_data= valid_spec_ds,
    epochs= EPOCHS,
)
```

```python
from tensorflow.python.ops.math_ops import arg_max
for (spectogram, label) in (test_spec_ds):
  pred= model.predict(spectogram)
  pred_label_index= arg_max(pred, dimension=1)
  orig_label= label_names[label]
  pred_label= label_names[pred_label_index]

  for j, label_ in enumerate(orig_label):
    print(f"Data- {j}: Original label {label_}, Prediction {pred_label[j]}")
```

```python
skor= model.evaluate(test_spec_ds)
skor[1]
```

```python
metrics= history.history

plt.plot(history.epoch, 100*np.array(metrics['accuracy']), 100*np.array(metrics['val_accuracy']))
plt.legend(['accuracy', 'val_accuracy'])

plt.ylim([0,100])
plt.xlabel('epoch')
plt.ylabel('accuracy[%]')
```
